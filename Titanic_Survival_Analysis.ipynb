Part 2: Clean Data [20 marks]
Tasks:
1. Check missing values
2. Fill Age with mean
3. Fill Embarked with mode
4. Drop Cabin column
5. Convert Sex to numeric (male=0, female=1)
Questions:
Which column had most missing values?
Why drop Cabin?

```python
#Check missing values

print(df.isnull().sum())
```

```python
# Fill Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'].isnull().sum())
```

```python
#Fill Embarked with mode

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.isnull().sum())
```

```python
#Drop Cabin column

df3 = df.drop(columns = ["Cabin"])

for columns in df3.columns:
    print(columns)
```

```python
#Convert Sex to numeric (male=0, female=1)

df3['Sex'] =  df3['Sex'].map({"male":0, "female":1})
df3['Sex']
```

```python
#Which column had most missing values?

most_missing = df.isnull().sum()
print(most_missing[most_missing > 0])
```

```python
#Why drop Cabin?



missing_percent = (df['Cabin'].isnull().sum() / len(df)) * 100
print(f"Cabin missing: {missing_percent:.2f}%")
print ("Too Many Missing Values")
```

🔄 Part 3: Create New Features [20 marks]
Tasks:
1. Create FamilySize = SibSp + Parch
2. Create IsAlone = 1 if FamilySize = 0 else 0
3. Extract Title from Name (Mr, Mrs, Miss, etc.)

Questions:
What does FamilySize tell us?
How many passengers were alone?

```python
#Create FamilySize = SibSp + Parch


df3['FamilySize'] = df3['SibSp'] + df3['Parch']
df3
```

```python
#Create IsAlone = 1 if FamilySize = 0 else 0

df3['IsAlone'] = df3['FamilySize'].apply(lambda x: 1 if x == 0 else 0)
df3
```

```python
#Extract Title from Name (Mr, Mrs, Miss, etc.)

df3['Title'] = df3['Name'].str.extract(r',\s*([^\.]+)\.')
print(df3['Title'].value_counts())
```

```python
#What does FamilySize tell us?
df3['FamilySize'] = df3['SibSp'] + df3['Parch']
```

```python
#How many passengers were alone?


import matplotlib.pyplot as plt

counts = df3['IsAlone'].value_counts()

plt.pie(counts,
        labels=['Not Alone', 'Alone'],
        autopct='%1.1f%%',
        colors=['skyblue', 'salmon'])

plt.title('Passengers Alone vs With Family')
plt.show()
```

📊 Part 4: Explore Data [20 marks]
Tasks:
1. Survival count
2. Survival rate by gender
3. Survival rate by passenger class
4. Average age of survivors vs non-survivors


Questions:
Which gender survived more?
Which class had highest survival?
Does age affect survival?

```python
#Survival count

(df3['Survived'].value_counts())
```

```python
#Survival rate by gender


survival_by_gender = df3.groupby('Sex')['Survived'].mean()

ax = survival_by_gender.plot(kind='bar',
                        color=['skyblue', 'salmon'],
                        edgecolor='black')

for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f'{bar.get_height():.1%}',
            ha='center', fontsize=12)

plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=0)
plt.ylim(0, 1)
plt.show()
```

```python
#Survival rate by passenger class

survival_by_class = df3.groupby('Pclass')['Survived'].mean()

ax = survival_by_class.plot(kind='bar',
                        color=['gold', 'skyblue', 'salmon'],
                        edgecolor='black')


for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f'{bar.get_height():.1%}',
            ha='center', fontsize=12)

plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0,1,2], labels=['1st','2nd','3rd'], rotation=0)
plt.ylim(0, 1)
plt.show()
```

```python
#Average age of survivors vs non-survivors

avg_age = df3.groupby('Survived')['Age'].mean()
print(avg_age)

ax = avg_age.plot(kind='bar',
                  color=['salmon', 'skyblue'],
                  edgecolor='black')


for bar in ax.patches:
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.3,
            f'{bar.get_height():.1f}',
            ha='center', fontsize=12)

plt.title('Average Age of Survivors vs Non-Survivors')
plt.xlabel('Survived')
plt.ylabel('Average Age')
plt.xticks(ticks=[0,1], labels=['Not Survived','Survived'], rotation=0)
plt.ylim(0, 50)
plt.show()
```


Questions:

Which gender survived more? Answer : Female

Which class had highest survival? Answer : 1st

Does age affect survival? Answer : Yes

```python
#Find correlation between numeric features

corr = df3.select_dtypes(include='number').corr()
print(corr)

plt.figure(figsize=(10, 6))
sns.heatmap(corr,
            annot=True,
            fmt='.2f',
            cmap='coolwarm')

plt.title('Correlation Between Numeric Features')
plt.show()
```

```python
#Find top 5 oldest passengers

top5_oldest = df3.sort_values('Age', ascending=False).head(5)
print(top5_oldest[['Name','Age']])
```

```python
#Fnd passenger with highest fare

highest_fare = df3['Fare'].idxmax()
print(df3.loc[highest_fare, ['Name', 'Fare', 'Sex', 'Pclass', 'Survived']])
```
